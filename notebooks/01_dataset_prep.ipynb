```python
# Automatic first tagging based on Medicinal Properties

tag_rules = {
    "digestion": ["stomachic", "digestive", "carminative", "apperizer", "laxative"],
    "pain": ["analgesic", "anodyne", "odontalgic", "antirheumatic"],
    "inflammatory": ["antiinflammatory", "antiphlogistic"],
    "respiratory_support": ["antiasthmatic", "expectorant", "pectorale"],
    "immune_support": ["antibacterial", "antibiotic", "antifungal", "antiseptic", "antiviral"],
    "skin": ["emollient", "vulnerary", "poultice", "desinfectant", "rubefacient"],
    "stress": ["sedative", "nervine", "antidepressant", "antianxiety"],
    "sleep": ["sedative", "nervine"],
    "anxiety": ["antianxiety", "nervine", "sedative"],
    "relaxation": ["sedative", "nervine"],
    "heart_support": ["cardiac", "vasodilator", "hypotensive"],
    "circulation": ["vasodilator", "diaphoretic"],
    "edema": ["diuretic"],
    "energy": ["stimulant", "tonic"],
    "focus": ["stimulant", "tonic"]
}

def generate_user_tags(properties):
    if pd.isna(properties):
        return ""
    
    text = properties.lower()
    tags = []

    for user_tag, keywords in tag_rules.items():
        if any(keyword in text for keyword in keywords):
            tags.append(user_tag)
    
    return ",".join(sorted(set(tags)))


def clean_effect_tags(properties):
    if pd.isna(properties):
        return ""
    
    return properties.lower().replace("; ", ", ").replace(";", ",")


# Apply to MVP data
mvp_data["user_need_tags"] = mvp_data["Medicinal Properties"].apply(generate_user_tags)
mvp_data["effect_tags"] = mvp_data["Medicinal Properties"].apply(clean_effect_tags)
```

## 5. Adding: Warning Level

```python
def assign_warning_level(hazard):
    if pd.isna(hazard) or hazard.lower() == "none known":
        return "low"
    
    text = hazard.lower()

    high_keywords = ["poisonous", "toxic", "pregnant", "abortion", "death", "convulsions"]
    medium_keywords = ["allergic", "irritation", "dermatitis", "sensitization", "caution"]

    if any(word in text for word in high_keywords):
        return "high"
    elif any(word in text for word in medium_keywords):
        return "medium"
    else:
        return "medium"
    
mvp_dataset["warning_level"] = mvp_dataset["Known Hazards"].apply(assign_warning_level)
```

```python
danger_keywords = [
    "poisonous",
    "toxic",
    "should not be used",
    "avoid",
    "death",
    "convulsions",
    "abortion",
    "pregnant",
    "pragnancy",
    "breastfeeding"
    "contraindicated",
    "cardiogenic",
    "highly irritant",
    "narcotic"
    "kidney",
    "liver",
    "blood pressure",
    "heart",
    "medication",
    "interaction",
    "large doses",
    "overdose",
    "vomitting",
    "seizures"
]

def has_high_risk_warning(row):
    hazard_text = str(row.get("Known Hazards", "")).lower()
    return any(keyword in hazard_text for keyword in danger_keywords)
```

```python
# Improved preparation type assignment
preparation_rules = {
    "tea": ["tea", "infusion", "decoction", "herbal tea"],
    "oil": ["oil", "essential oil", "aromatherapy"],
    "topical": ["topical", "skin", "wound", "burn", "external", "externally"],
    "cream_or_onitment": ["cream", "onitment", "salve", "balm", "lotion"],
    "poultice": ["compress", "plaster"],
    "extract": ["extract", "tincture", "essence"],
    "capsule_or_powder": ["capsule", "powder", "dried", "ground"],
    "juice_or_gel": ["juice", "gel", "sap"],
    "food": ["edible", "eaten", "culinary", "spice", "flavouring", "salad"],
    "smoke_or_incense": ["smoke", "smoking", "incense", "fumigation", "burned", "burnt"]
}
```

```python
def assign_preparation_type(row):
    # safety override first
    if has_high_risk_warning(row):
        return "not recommended due to safety concerns"
    
    text = " ".join([
        str(row.get("Medicinal Properties", "")),
        str(row.get("Summary", "")),
        str(row.get("Known Hazards", "")),
        str(row.get("effect_tags", "")),
        str(row.get("user_need_tags", ""))
    ]).lower()

    found_types = []

    for prep_type, keywords in preparation_rules.items():
        if any(keyword in text for keyword in keywords):
            found_types.append(prep_type)

    if not found_types:
        return "unknown"
    
    return ",".join(sorted(set(found_types)))
```

```python
mvp_dataset["preparation_type"] = mvp_dataset.apply(
    assign_preparation_type,
    axis=1
)
```
